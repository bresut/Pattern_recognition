```python
x,y = symbols('x , y')
x1 = exp(-((x-1)**2+(y+1)**2))
x1.diff(x)
```

```python
x1.diff(y)
```

```python
xa = x1.diff(x)
ya = x1.diff(y)
eq1 = sym.Eq(xa,0)
eq2 = sym.Eq(ya,0)
eq1
```

```python
sym.solve([eq1, eq2],x,y,  

         dict=True,  

         simplify=True 

        ) 
```

```python
q= sym.plotting.plot3d(x1, 

                      (x,-10,10),(y,-10,10), 

                      xlabel='x', 

                      ylabel='y', 

                      title=f'z= {x1}') 
```
